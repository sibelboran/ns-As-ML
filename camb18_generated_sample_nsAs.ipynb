# directory and path

```python
## directory
directory = 'planck18_CMB_ps_dataset/'
```

```python
planck18_fname = 'COM_PowerSpect_CMB-base-plikHM-TTTEEE-lowl-lowE-lensing-minimum-theory_R3.01.txt'
planck18       = np.loadtxt(directory + planck18_fname)
```

```python
calPlanck = 0.1000442E+01
l         = planck18[:,0]
TT        = planck18[:,1] / calPlanck**2
TE        = planck18[:,2] / calPlanck**2
EE        = planck18[:,3] / calPlanck**2
```

```python
pars    = camb.CAMBparams()
pars.set_cosmology(H0=0.6732117E+02, ombh2=0.2238280E-01, omch2=0.1201075E+00, mnu=0.06, omk=0, tau=0.5430842E-01)
pars.InitPower.set_params(ns=0.9660499E+00, As=2.100549e-9, r=0)
pars.set_for_lmax(int(max(l)), lens_potential_accuracy=0);
```

```python
results = camb.get_results(pars);
```

```python
ti      = time()
camb_TT = results.get_cmb_power_spectra(pars, CMB_unit='muK')['total'][2:len(l)+2, 0]
for name in camb_TT: print(name)
tf      = time()
print((tf-ti)/60, 'mins')
```

```python
print('camb_TT.size:',  camb_TT.size)
print('camb_TT.dtype:', camb_TT.dtype)
print('camb_TT:',       camb_TT)
```

```python
plt.xlabel('$\ell$')
plt.ylabel('$\mathcal{D}_{\ell}^{TT}$')
plt.title('TT')
plt.plot(TT, '.', color='red',  label='planck18_TT')
plt.plot(camb_TT, color='blue', label='camb_TT')
plt.legend()
```

```python
ti      = time()
camb_TE = results.get_cmb_power_spectra(pars, CMB_unit='muK')['total'][2:len(l)+2, 3]
for name in camb_TE: print(name)
tf      = time()
print((tf-ti)/60, 'mins')
```

```python
print('camb_TE.size:',  camb_TE.size)
print('camb_TE.dtype:', camb_TE.dtype)
print('camb_TE:',       camb_TE)
```

```python
plt.xlabel('$\ell$')
plt.ylabel('$\mathcal{D}_{\ell}^{TE}$')
plt.title('TE')
plt.plot(TE, '.', color='red',  label='planck18_TE')
plt.plot(camb_TE, color='blue', label='camb_TE')
plt.legend()
```

```python
ti      = time()
camb_EE = results.get_cmb_power_spectra(pars, CMB_unit='muK')['total'][2:len(l)+2, 1]
for name in camb_EE: print(name)
tf      = time()
print((tf-ti)/60, 'mins')
```

```python
real_ns = 0.9660499E+00
real_As = 2.100549e-9

n_data  = 10000

ns_list = np.zeros([n_data, 1])
As_list = np.zeros([n_data, 1])

for i in range(n_data):
    ns_list[i] = real_ns + np.random.randn()*(real_ns/10)
    As_list[i] = real_As + np.random.randn()*(real_As/10)
    
np.savetxt('camb18_dataset/generated_camb18_dataset/generated_ns_list.txt', ns_list)
np.savetxt('camb18_dataset/generated_camb18_dataset/generated_As_list.txt', As_list)
```

```python
# np.random.seed(1)
n_data         = 1000
generated_data = np.zeros([n_data, 2+len(TT)])

ti     = time()

for i in range(n_data):

    pars = camb.CAMBparams()
    pars.set_cosmology(H0=0.6732117E+02, ombh2=0.2238280E-01, omch2=0.1201075E+00, mnu=0.06, omk=0.0, tau=0.5430842E-01)
    pars.InitPower.set_params(ns=ns_list[i], As=As_list[i], r=0)
    pars.set_for_lmax(int(max(l)), lens_potential_accuracy=0);
    
    generated_data[i,0]  = real_ns + np.random.randn()*(real_ns/10)
    generated_data[i,1]  = real_As + np.random.randn()*(real_As/10)
    generated_data[i,2:] = results.get_cmb_power_spectra(pars, CMB_unit='muK')['total'][2:len(l)+2, 0]
    
    if i%100 == 0:
        print(i+100, (time()-ti)/60, 'mins')
        np.savetxt('camb18_dataset/generated_camb18_dataset/generated_sample_nsAs.txt', generated_data)
```
