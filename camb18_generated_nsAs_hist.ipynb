```python
ns_As   = []
mean_ns = 0.96
mean_As = 3.04
std     = 0.3
np.random.seed(189921)

while (len(ns_As)<=10000):
    ns = np.random.uniform(0.55, 1.37)
    As = np.random.uniform(2.68, 3.5)
    
    if ns>0.1 and ns<1.82 and As<5.09 and As>1.09:
        ns_As.append([ns,As])

ns_As = np.array(ns_As)
cp    = ns_As
```

```python
cp_train = cp[1000:]
cp_test  = cp[:1000]
```

```python
f, ax = plt.subplots(2, 1, figsize=(2, 4), dpi=300)

fs = 4

ax[0].scatter(cp_train[:,0], cp_train[:,1], s=.1, color='pink',   label='Train')
ax[0].scatter(cp_test[:,0],  cp_test[:,1],  s=.2, color='fuchsia', label='Test')
ax[0].scatter([0.9649], [3.045], color='lime',s=9.0)
ax[0].set_xlabel('$\mathcal{n}_s$', labelpad=0.8, fontsize=5)
ax[0].set_ylabel('$Log(10^{10}\mathcal{A}_s)$',   fontsize=5)
ax[0].tick_params(axis='both',   labelsize=3.5)
ax[0].legend(loc='upper center', fontsize=fs)

ax[1].hist([cp_train[:,0], cp_test[:,0]], bins=10, stacked=True, color=['pink', 'fuchsia'], label=['Train', 'Test'], histtype='stepfilled')
ax[1].hist([cp_train[:,1], cp_test[:,1]], bins=10, stacked=True, color=['pink', 'fuchsia'], histtype='stepfilled')
ax[1].set_xticks([0.75, 1.00, 1.25, 2.50, 3.00, 3.50])   
ax[1].set_yticks([200, 400, 600, 800, 1000]) 
ax[1].text(0.15, -0.18, '$\mathcal{n}_s$', transform=ax[1].transAxes, fontsize=5)
ax[1].text(0.65, -0.18, '$Log(10^{10}\mathcal{A}_s)$', transform=ax[1].transAxes, fontsize=5)
ax[1].axvline(0.9649, color='lime', lw=0.75)
ax[1].axvline(3.045,  color='lime', lw=0.75)
ax[1].tick_params(axis='both', labelsize=3.5)
ax[1].set_ylabel('Number of samples', fontsize=5)
ax[1].legend(loc='upper center', fancybox=False, framealpha=0.3, fontsize=fs)

#plt.savefig('camb18_generated_ns_As_hist.pdf', bbox_inches='tight', dpi=300)
```
